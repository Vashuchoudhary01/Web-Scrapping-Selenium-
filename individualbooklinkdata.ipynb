```python
import pandas as pd
from bs4 import BeautifulSoup 
from tqdm import tqdm 
import requests as rq
```

```python
df = pd.read_csv("BookScrapping.csv")
```

```python
data =  []
for link in tqdm(df["book_link"]):
    res = rq.get(link)
    soup = BeautifulSoup(res.text, "html.parser")
    # take the type of it
    typ = soup.find("ul", class_ = "breadcrumb").find_all("a")[2].text
   
    UPC          = soup.find("table", class_ = "table table-striped").find_all("td")[0].text
    Price_el_tx  = soup.find("table", class_ = "table table-striped").find_all("td")[2].text[2:]
    Price_in_tx  = soup.find("table", class_ = "table table-striped").find_all("td")[3].text[2:]
    Tax          = soup.find("table", class_ = "table table-striped").find_all("td")[4].text[2:]
    qnty         = soup.find("table", class_ = "table table-striped").find_all("td")[5].text
    reviews      = soup.find("table", class_ = "table table-striped").find_all("td")[6].text
  
    data.append([UPC, Price_el_tx, Price_in_tx, Tax, qnty, reviews])
```

```python
df = pd.DataFrame(data, columns = ["UPC", "Price_el_tx", "Price_in_tx", "Tax", "qnty", "reviews"])
df
```

```python
df.to_csv("Scrabbookpage.csv", index = False)
```
