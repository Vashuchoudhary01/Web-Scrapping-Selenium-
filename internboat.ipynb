```python
last_height = driver.execute_script("return document.body.scrollHeight")

while True:
    # Scroll down
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")

    # Wait for new content to load
    time.sleep(2)  

    # Calculate new scroll height
    new_height = driver.execute_script("return document.body.scrollHeight")
    
    # Check if we reached the bottom
    if new_height == last_height:
        break
    
    last_height = new_height
driver.execute_script("window.scrollTo(0, 0);")
print("Page fully loaded")
```

```python
all_intern = driver.find_element(By.CLASS_NAME, "internship_list_container")

internships = all_intern.find_elements(By.CSS_SELECTOR, ".container-fluid.individual_internship.view_detail_button.visibilityTrackerItem") 
time.sleep(3)
len(internships)
```

```python
# this is for get the text data and store it
data = []
for intern in tqdm(internships):
    profile_link = intern.find_element(By.TAG_NAME,"div").find_element(By.CLASS_NAME, "job-title-href").get_attribute("href")
    profile_name = intern.find_element(By.TAG_NAME,"div").find_element(By.CLASS_NAME, "job-title-href").text 
    cmp_nn = intern.find_element(By.CLASS_NAME,"company-name").text
    details_span = intern.find_element(By.CLASS_NAME, "detail-row-1").text.split("\n")
    location = details_span[0]
    Duration = details_span[1]
    stipend = details_span[2]
    try:
        hiring_acitivity = intern.find_element(By.CLASS_NAME, "actively-hiring-badge").text
        # print(hiring_acitivity)
        
        
    except:
        hiring_acitivity= "Not Mentioned"
        
    data.append([profile_name,profile_link,cmp_nn,hiring_acitivity,location,Duration,stipend])   
```

```python
import pandas as pd
df = pd.DataFrame(data, columns = ["profile_name","profile_link","cmp_nn","hiring_acitivity","location","Duration","stipend"])
df
```

```python
for intern in internships:
    # Open the link in a new tab (Uncomment if needed)
    profile_link_element = intern.find_element(By.TAG_NAME, "div").find_element(By.CLASS_NAME, "job-title-href")
    profile_link_element.send_keys(Keys.CONTROL + Keys.RETURN)
    print("new window is open for application")
    time.sleep(2)
    
    WebDriverWait(driver, 5).until(lambda d: len(d.window_handles) > 1)
    windows = driver.window_handles  
    if len(windows) > 1:
        new_window = windows[-1]  
        driver.switch_to.window(new_window)
    else:
        print("New window did not open.")
        continue  # Move to the next internship
    
    time.sleep(2)
    
    print(driver.current_url)
    try:
        apply_button_for_jbs = driver.find_element(By.CLASS_NAME, "detail_view").find_element(By.CLASS_NAME, "buttons_container")
        apply_button_for_jbs.click()
        time.sleep(3)
        
        proceed_link = driver.find_element(By.CSS_SELECTOR, ".btn.btn-large.education_incomplete.proceed-btn")
        proceed_link.click()
        time.sleep(3)
        WebDriverWait(driver, 5).until(lambda d: len(d.window_handles) > 1)
        
        windows = driver.window_handles  
        new_window = windows[-1]  
        driver.switch_to.window(new_window)
    except NoSuchElementException:
        print("Apply or proceed button not found. Skipping...")

    assignment_qn = driver.find_element(By.ID, "assessment_questions")

    try:
        qn = assignment_qn.find_element(By.CLASS_NAME, "assessment_question") 
    
        answer = assignment_qn.find_element(By.CSS_SELECTOR, ".ql-editor.ql-blank")  

        fill_answer = "You should hire me because I have a strong foundation in Data Science, AI, Machine Learning, and Selenium, which allows me to handle data analysis, build predictive models, and automate processes efficiently. I have hands-on experience in web scraping, supervised learning, and automation, which makes me a versatile candidate. I am a quick learner, problem-solver, and passionate about using data-driven solutions to solve real-world problems. My ability to combine analytics, automation, and machine learning makes me a great fit for this role."

        answer.send_keys(fill_answer)  
        answer.send_keys(Keys.RETURN)  
        print("Question found, answer filled.")
    except:
        print("Question not found")

    try:
        final_sub_button = assignment_qn.find_element(By.CLASS_NAME, "submit_button_container")
        final_sub_button.click()
        print("Submit button clicked.")
        try:
            WebDriverWait(driver, 5).until(lambda d: len(d.window_handles) > 1)
            windows = driver.window_handles  
            new_window = windows[-1]  
            driver.switch_to.window(new_window)
        except:
            print("we are on the last window")
    except NoSuchElementException:
        print("Submit button not found.")

    try:
        continue_apply = driver.find_element(By.CLASS_NAME, "already_applied_text")
        move_frd = continue_apply.find_element(By.TAG_NAME, "a").get_attribute("href")
        driver.get(move_frd)
        try:
            WebDriverWait(driver, 5).until(lambda d: len(d.window_handles) > 1)
            windows = driver.window_handles  
            new_window = windows[-1]  
            driver.switch_to.window(new_window)
        except:
            print("we are on the last window")
          # Stop after the first internship
    except NoSuchElementException:
        print("No such element found.")
```
