```python
import requests as rq
import pandas as pd
from bs4 import BeautifulSoup
from tqdm import tqdm
```

```python
# now i am going to visit all the pages here it sself
data = []
for page in tqdm(range(1,11)):
    a= "https://quotes.toscrape.com/page/" + str(page)
    res = rq.get(a)
    soup = BeautifulSoup(res.text,"html.parser")
    # now do the same work but for the different pages
    for sp in soup.find_all("div", class_="quote"):
        quotes = sp.find("span", class_="text").text[1:-1]
        authors_name= sp.find("small", class_="author").text
        author_id= sp.find("a").get("href")
        tags = []
        for tag in sp.find_all("a", class_ ="tag"):
            tags.append(tag.text)
        tags = ",".join(tags)
    # now combine all the things inside a single dataset
        data.append([quotes,authors_name,author_id, tags])
    
```

```python
len(data)
```

```python
df2 = pd.DataFrame(data, columns=["quotes","authors_name","authors_id","tags"])
```
